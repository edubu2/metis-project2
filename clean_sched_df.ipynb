```python
mask = (sched_df.team == 'chi') & (sched_df.year == 2019)
sched_df[mask]
```

```python
# clean the game_location column & apply change. def home_game(row):
def home_game(row):    
    if row == '@':
        return 1
    return 0

sched_df['home'] = sched_df.game_location.apply(home_game)
sched_df.drop("game_location", axis=1, errors='ignore', inplace=True)
```

```python
sched_df[mask]
```

```python
# add team_year identifier to each row (to easily groupby team/year)

team_years = sched_df['team'] + '-' +  sched_df['year'].astype(str) 
sched_df.insert(loc=2, column='team_year', value=team_years)
sched_df
```
