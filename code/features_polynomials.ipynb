```python
# create squared features

ewma_cols = [col for col in game_df.columns if col[:4] == 'ewma']
ewma_cols.insert(0, 'year')

ewma_cols.append('roll3_num_wins')
ewma_cols.append('roll3_num_wins_opp')
ewma_cols.append('roll16_num_wins')
ewma_cols.append('roll16_num_wins')

for ewma in ewma_cols:
    new_col = 'p2_' + ewma
    try:
        game_df[new_col] = game_df[ewma] ** 2
    except:
        print(new_col + ' failed.')
```

```python
features = ['year', 'team_home_game']

p2_cols = [col for col in game_df.columns if col[:3] == 'p2_']

for col in p2_cols:
    features.append(col)

X = game_df[features]

y = game_df['margin']

model = split_and_validate2(X, y, print_results=True)
```
